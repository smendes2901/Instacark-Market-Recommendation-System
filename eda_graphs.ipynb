```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
order_products_prior = pd.read_csv("datasets/order_products__prior.csv")
order_products_train  = pd.read_csv("datasets/order_products__train.csv")
```

```python
len(order_products_prior["order_id"].unique())
```

```python
len(order_products_train["order_id"].unique())
```

```python
set(order_products_train["order_id"].unique()).intersection(order_products_prior["order_id"].unique())
```

```python
order_products = pd.concat([order_products_prior, order_products_train], axis=0, ignore_index=True)
```

```python
order_products.head()
```

```python
product_details  = pd.read_csv("datasets/product_details.csv")
product_details.head()
```

```python
product_details.shape
```

```python
len(product_details["product_id"].unique())
```

```python
order_product_details = order_products.merge(product_details, on="product_id")
order_product_details.columns
```

```python
order_product_details.to_csv("datasets/order_product_details.csv", index=False)
```

```python
order_product_details.department.value_counts().sort_values().plot(kind = 'barh')
```
